```
import requests
import json

page=requests.get("https://api.datamuse.com/words?rel_rhy=fun")
print(type(page))
print(page.text[:150])
print(page.url)
print("-----")
x=page.json()
print(type(x))
print('-----')
print(x[0])
print(json.dumps(x, indent=2))
```
